```python
import pandas as pd

#--- GET THE ORDERS DATASET
orders_df = pd.read_csv("./order_details.csv")
```

```python
orders_df.head()
```

```python
#dataset info
orders_df.info()
```

```python
#--- CLEAN THE ORDERS DATASET
#will drop any row with missing values
orders_df = orders_df.dropna()
```

```python
#--- JOIN THE ORDERS DATASET WITH THE MENU ITEMS DATASET
menu_df = pd.read_csv("./menu_items.csv")
menu_df.info()
```

```python
#join/merge the datasets - and remove the extra id
order_items_df = orders_df.merge(menu_df, how="left", left_on="item_id", right_on="menu_item_id").drop("menu_item_id", axis=1)
order_items_df.head()
```

```python
#--- ADD TAXES AND TOTAL REVENUE COLUMNS TO THE NEW MERGED DATASET
#taxes (get the price column and multiply the tax amount and rounde it)
order_items_df["sales_tax"] = (order_items_df.price * 0.08).round(2)

#total revenue
order_items_df["total_renevue"] = order_items_df.price + order_items_df.sales_tax 

order_items_df.head()
```

```python
order_items_df.describe()
```

```python
#--- GET THE SUM OF REVENUE FOR ALL THE ITEMS
#group by item_name column and sum the total_revenue - and show it in a bar graph
order_items_df.groupby("item_name").agg({"total_renevue": "sum"}).sort_values("total_renevue").plot.barh()
```

```python
#can also filter data by category column
order_items_df.query("category == 'Italian'").groupby("item_name").agg({"total_renevue": "sum"}).sort_values("total_renevue").plot.barh()
```
